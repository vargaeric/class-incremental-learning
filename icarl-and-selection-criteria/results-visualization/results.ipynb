```python
import matplotlib.pyplot as plt

def convert_to_percentage_accuracies(input_list):
    return [element * 100 for element in input_list]

def create_plot(original_accuracies, median_accuracies, kmeans_accuracies,
                density_accuracies, classes_number):
  plt.plot(classes_nr, convert_to_percentage_accuracies(original_accuracies), marker='o', color='blue', label='Original')
  plt.plot(classes_nr, convert_to_percentage_accuracies(median_accuracies), marker='^', color='green', label='Median')
  plt.plot(classes_nr, convert_to_percentage_accuracies(kmeans_accuracies), marker='s', color='red', label='K-Means')
  plt.plot(classes_nr, convert_to_percentage_accuracies(density_accuracies), marker='x', color='purple', label='Density')
  plt.xlabel('Number of Classes')
  plt.ylabel('Accuracy')
  plt.xticks(classes_nr)
  plt.legend()
  plt.grid(True)
  plt.show()

def calculate_average_differences(original, compared):
    differences = [c - o for o, c in zip(original, compared)]
    average_difference = sum(differences) / len(differences)

    return average_difference

def get_avg_differences(original_accuracies, median_accuracies,
                        kmeans_accuracies, density_accuracies):
  avg_diff_original_median = calculate_average_differences(original_accuracies, median_accuracies) * 100
  avg_diff_original_kmeans = calculate_average_differences(original_accuracies, kmeans_accuracies) * 100
  avg_diff_original_density = calculate_average_differences(original_accuracies, density_accuracies) * 100

  print()
  print("Average difference between original and median:", avg_diff_original_median)
  print("Average difference between original and kmeans:", avg_diff_original_kmeans)
  print("Average difference between original and density:", avg_diff_original_density)
```
