```python
import matplotlib.pyplot as plt

markers = ['-o', '-^', '-s', '-x', '-v']
colors = ['tab:blue', 'tab:green', 'tab:red', 'tab:purple', 'tab:orange']

def create_partial_plots(classes_or_task_numbers, values, labels, ylabel, xlabel, title):
  plt.grid(True)

  for i in range(len(labels)):
    plt.plot(classes_or_task_numbers[i], values[i], markers[i], label=labels[i], color=colors[i])

  plt.ylabel(ylabel)
  plt.xlabel(xlabel)
  plt.title(title)
  plt.legend()
  plt.xticks(classes_or_task_numbers[0])

def plot(classes_or_task_numbers, first_values, second_values, labels, xlabel):
  plt.figure(figsize=(10, 5))
  plt.subplot(1, 2, 1)
  create_partial_plots(classes_or_task_numbers, first_values, labels, 'Accuracy', xlabel, 'Accuracy Comparison')
  plt.subplot(1, 2, 2)
  create_partial_plots(classes_or_task_numbers, second_values, labels, 'Forgetting', xlabel, 'Forgetting Comparison')
  plt.tight_layout()
  plt.show()
```
